```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/drive/MyDrive/heart.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
df.describe()
```

```python
#target distribution
df.target.value_counts()
```

```python
#shoeing correlation matrix
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#plot target district
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['not disease','disease'])
plt.title('Target Distribution');
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA53B','red'])
plt.legend("haven't disease","have disease")
plt.title('Heart disease Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#sns.set_style("whitegrid")
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']== 0],bins=30,color="green",label='have disease')
sns.distplot(df.age[df['target']== 1],bins=30,color="red",label='have not disease')
plt.legend()
plt.title('Heart Disease Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
